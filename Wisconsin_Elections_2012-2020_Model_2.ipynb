```python
# Extract US House Election Data
districtResults = elect_201216[['CON', 'USHTOT16', 'USHDEM16', 'USHREP16', 'USHTOT14', 'USHDEM14', 'USHREP14', 'USHTOT12', 'USHDEM12', 'USHREP12']]
```

```python
votes_districtByYear = pd.DataFrame(columns = ['district', 'TOT12', 'DEM12', 'REP12', 'TOT14', 'DEM14', 'REP14', 'TOT16', 'DEM16', 'REP16'])
```

```python
# Wisconsin has 8 congressional districts
districtList = [1,2,3,4,5,6,7,8]

# Sum total democratic and republican votes in each district for each year
for district in districtList:
    tot_votes_12 = 0
    dem_votes_12 = 0
    rep_votes_12 = 0
    tot_votes_14 = 0
    dem_votes_14 = 0
    rep_votes_14 = 0
    tot_votes_16 = 0
    dem_votes_16 = 0
    rep_votes_16 = 0
    
    for row in range(len(districtResults)):
        if(int(districtResults.iloc[row]['CON']) == int(district)):
            tot_votes_12 = tot_votes_12 + int(districtResults.iloc[row]['USHTOT12'])
            dem_votes_12 = dem_votes_12 + int(districtResults.iloc[row]['USHDEM12'])
            rep_votes_12 = rep_votes_12 + int(districtResults.iloc[row]['USHREP12'])
            tot_votes_14 = tot_votes_14 + int(districtResults.iloc[row]['USHTOT14'])
            dem_votes_14 = dem_votes_14 + int(districtResults.iloc[row]['USHDEM14'])
            rep_votes_14 = rep_votes_14 + int(districtResults.iloc[row]['USHREP14'])
            tot_votes_16 = tot_votes_16 + int(districtResults.iloc[row]['USHTOT16'])
            dem_votes_16 = dem_votes_16 + int(districtResults.iloc[row]['USHDEM16'])
            rep_votes_16 = rep_votes_16 + int(districtResults.iloc[row]['USHREP16'])
    
    totalDistrict = [district, tot_votes_12, dem_votes_12, rep_votes_12, tot_votes_14, dem_votes_14, rep_votes_14, tot_votes_16, dem_votes_16, rep_votes_16]
    df_totalByDistrict = pd.DataFrame([totalDistrict], columns =  ['district', 'TOT12', 'DEM12', 'REP12', 'TOT14', 'DEM14', 'REP14', 'TOT16', 'DEM16', 'REP16'])
    votes_districtByYear = votes_districtByYear.append(df_totalByDistrict, ignore_index = True)
```

```python
votes_districtByYear
# Extremely low Number of voters in District 4 for 2016. This is due to the fact Milwaukee is contained in district 4 
# and many voters in this area voted for Green or Independent Party. A similar result occurred for District 3.
```

```python
dem_proportionByDistrict = votes_districtByYear.copy(deep=True)
```

```python
dem_districts12 = pd.DataFrame({'dem_prop_12':dem_proportionByDistrict['DEM12']/dem_proportionByDistrict['TOT12']})
dem_districts14 = pd.DataFrame({'dem_prop_14':dem_proportionByDistrict['DEM14']/dem_proportionByDistrict['TOT14']})
dem_districts16 = pd.DataFrame({'dem_prop_16':dem_proportionByDistrict['DEM16']/dem_proportionByDistrict['TOT16']})
```

```python
dem_proportionByDistrict = pd.concat([dem_proportionByDistrict, dem_districts12, dem_districts14, dem_districts16], axis=1)
```

```python
dem_proportionByDistrict
```

```python
# Tally wins for democrats in each district for each year. 0 if democrats win, 1 otherwise.

dem_win_12 = []
dem_win_14 = []
dem_win_16 = []

for row in range(len(votes_districtByYear)):
    if(votes_districtByYear.iloc[row]['DEM12'] > votes_districtByYear.iloc[row]['REP12']):
        dem_win_12.append(1)
    else:
        dem_win_12.append(0)
        
    if(votes_districtByYear.iloc[row]['DEM14'] > votes_districtByYear.iloc[row]['REP14']):
        dem_win_14.append(1)
    else:
        dem_win_14.append(0)
        
    if(votes_districtByYear.iloc[row]['DEM16'] > votes_districtByYear.iloc[row]['REP16']):
        dem_win_16.append(1)
    else:
        dem_win_16.append(0)
    

df_dem12 = pd.DataFrame({'DEM_WIN_12': dem_win_12})
df_dem14 = pd.DataFrame({'DEM_WIN_14': dem_win_14})
df_dem16 = pd.DataFrame({'DEM_WIN_16': dem_win_14})

votes_districtByYear = pd.concat([votes_districtByYear, df_dem12, df_dem14, df_dem16], axis=1)
    
```

```python
# Calculate average fraction of districts won by the Democratic Party
numberOfDistricts = 8
numberOfYears = 3

dem_totDistricts12 = 0
dem_totDistricts14 = 0
dem_totDistricts16 = 0


for row in range(len(votes_districtByYear)):
    dem_totDistricts12 = dem_totDistricts12 + votes_districtByYear.iloc[row]['DEM_WIN_12']
    dem_totDistricts14 = dem_totDistricts14 + votes_districtByYear.iloc[row]['DEM_WIN_14']
    dem_totDistricts16 = dem_totDistricts16 + votes_districtByYear.iloc[row]['DEM_WIN_16']

dem_win_frac12 = dem_totDistricts12/numberOfDistricts
dem_win_frac14 = dem_totDistricts14/numberOfDistricts
dem_win_frac16 = dem_totDistricts16/numberOfDistricts

avg_dem_winByDistricts = (dem_win_frac12 + dem_win_frac14 + dem_win_frac16)/numberOfYears    
# For given data, average fraction of districts won by Democratic Party is 0.375
```

```python
# Code by Tony S. Yu: https://tonysyu.github.io/plotting-error-bars.html#.WiCAL0qnFPZ
def errorfill(x, y, yerr, color=None, alpha_fill=0.3, ax=None):
    ax = ax if ax is not None else plt.gca()
    if color is None:
        #color = ax._get_lines.color_cycle.next()
        color = 'blue'
    if np.isscalar(yerr) or len(yerr) == len(y):
        ymin = y - yerr
        ymax = y + yerr
    elif len(yerr) == 2:
        ymin, ymax = yerr
    ax.plot(x, y, color=color)
    ax.fill_between(x, ymax, ymin, color=color, alpha=alpha_fill)
```
